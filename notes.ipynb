Machine Learning is a discipline that gathers set of structured and organized data
to generate/tune a set of rules/features that mold a system. This is used to 
either mimic a system, predict the behaviour under given circumstances, to clasify
a data record or to reduce the dimentionallity of a dataset.

Scikit-learn is a Machine Learning library that eases the process of building
Machine Learning models, pre process data, train models, validate models or update
models.

Note that Machine Learning consists in several steps:

1 - Data acquisition
2 - Data preprocessing
3 - Train
4 - Test

This package provides datasets to train different models and dive into the 
Machine Learning World

```python
# Exploring numeric and categorical features
df_d = pd.read_csv('resources/housing.csv')
df_d.head()
```

```python
# Linear Regression
# Note that whenever a ML model will be fitted there can be several values
# that seem to be exceptional values that may disturb the normal distribution 
# of the data. In such cases there can be excluded from the data to have a more
# accurate model. In this case there will be excluded all those house values that
# are 500001
df_f = df_e.drop(df_e.loc[df_e['median_house_value']==500001].index)
df_f.head()
```

```python
# Once data is clean there can be applied a method to get a set of categories 
# from a column. The resulting dataframe will suitable to provide data able to be
# numerically processed. This method is also called one hot encoder
print (f'df_f.shape before dummy variables columns declared {df_f.shape}')
df_f = pd.get_dummies(df_f, columns = ['ocean_proximity'])
print (f'df_f.shape after dummy variables columns declared {df_f.shape}')
df_f.head()
```

```python
# X and Y split
x_values_a = df_f.drop('median_house_value', axis = 1)
y_values_a = df_f['median_house_value']
```

```python
# The Machine Learning process requires to split data into training and testing. 
# Scikit-learn has a module to provide the splitted data in a simple manner. This
# principle applies for features and target values
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_values_a, y_values_a, test_size = 0.2)
print (f'x_train.shape, x_test.shape, y_train.shape, y_test.shape = {x_train.shape, x_test.shape, y_train.shape, y_test.shape}')
```

```python
# Now that data is splitted there can be trained a model
from sklearn.linear_model import LinearRegression

model_a = LinearRegression(normalize = True, fit_intercept = False).fit(x_train,y_train)
print (f'Train score: {model_a.score}') 
# The result is the r**2 value from the linear fitting
```

```python
# To check how a feature variable contributes to the target variable there's used
# the coef_ attribute

predictors_a = x_train.columns
coef_a = pd.Series(model_a.coef_,predictors_a).sort_values()
coef_a
```

```python
# The model performance can be assessed 
y_pred_a = model_a.predict(x_test)

df_h = pd.DataFrame({'Predicted': y_pred_a, 'Actual':y_test})
df_h.head()
```

```python
# Model performance assessment is made by employing the r**2 score method
from sklearn.metrics import r2_score

print (f'Testing score: {r2_score (y_test,y_pred_a)}')
```

```python
# There can be plotted the data
fig_e, ax_e = plt.subplots(figsize=(12,8))
plt.scatter(y_test,y_pred_a)
```

```python
# Logistic Regression
# Is a regression made to classify a record within a category. This leads to 
# a process in which a category for a record will be guessed under a percentage
# of probability that the result actually belongs to a either category.

# Logistic regression helps to find how posibilities are changed by action

# The Logistic regression problem will check if a price is over or below the 
# median price

median_value_a = df_f['median_house_value'].median()
```

```python
# Dataset processing
df_f['above_median'] = (df_f['median_house_value'] - median_value_a) > 0
```

```python
# X and Y split
x_values_b = df_f.drop(['median_house_value','above_median'], axis = 1)
y_values_b = df_f['above_median']
```

```python
# Train and test split
x_train, x_test, y_train, y_test = train_test_split(x_values_b, y_values_b, test_size = 0.2)
```

```python
# Model selection and fitting
from sklearn.linear_model import LogisticRegression
model_b = LogisticRegression(solver = 'liblinear').fit(x_train,y_train)
```

```python
# Model assessment
print (f'Train score: {model_b.score(x_train,y_train)}') 

y_pred_b = model_b.predict(x_test)

df_m = pd.DataFrame({'Predicted': y_pred_b, 'Actual':y_test})
df_m.head()

from sklearn.metrics import accuracy_score

print (f'Test score: {accuracy_score(y_test,y_pred_b)}') 
```
